```python
import pandas as pd
import numpy as np

##Visualization Libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<strong>Problem 1</strong>

```python
data = pd.read_csv('student_class.csv')
majors = data['Majors'].unique()
midterms = data.groupby(['Majors']).agg({'Midterm':'median'})
final = data.groupby(['Majors']).agg({'Final':'mean'})
avg = data.groupby(['Majors']).agg({'Final Average':'mean'})

label = np.arange(majors.shape[0])
w = 0.3
plt.bar(label-w/2, midterms['Midterm'], label='Midterm', width = w/2, color='red')
plt.bar(label, final['Final'], label='Final', width = w/2,color='green')
plt.bar(label+w/2, avg['Final Average'], label='Final Average', width = w/2,color='blue')
plt.xticks(label, labels=majors)
plt.gcf().set_size_inches(12, 5)
plt.xlabel('Major')
plt.ylabel('Median Score')
plt.legend(loc='upper right', prop={'size': 7})
plt.title('Median Midterm, Final, and Final Average Scores')
plt.gcf().autofmt_xdate()
plt.show()
```

Looking at the graph, BNFO students can be seen doing well during the midterm. In terms of the final, BNFO and IS students did well on that. For the Final Average, IT and BNFO students did well. After analyzing all that, one can see that <strong>BNFO</strong> students did the best overall throughout the semester and at the end.
